```python
ax1=plt.subplot(111)
pb1=pd.read_csv("DJIA.csv")

pb1=pb1[pb1.VALUE!="."]

pb1.VALUE=pd.to_numeric(pb1.VALUE)
pb1.DATE=pd.to_datetime(pb1.DATE)
ax1=plt.subplot(111)
pb1.plot(x='DATE',y='VALUE',ax=ax1)
pb1=pd.read_csv("DJIA.csv")
pb1=pb1[pb1.VALUE!="."]
pb1.VALUE=pd.to_numeric(pb1.VALUE)
pb1.rolling(window=10).mean().plot(color='g',ax=ax1)
pb1.rolling(window=10).min().plot(color='r',ax=ax1)
pb1.rolling(window=10).max().plot(color='b',ax=ax1)
pb1.ewm(span=10).mean().plot(color='r',ax=ax1)
ax1.legend(["actual","rolling av","rolling min","rolling max","exp av"])
```
